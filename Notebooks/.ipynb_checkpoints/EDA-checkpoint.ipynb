# Task description

In this competition we must predict 2 cases:
1. Predict class(one in a 4) of desease(classes is mutually exclusive, www.kaggle.com/c/siim-covid19-detection/discussion/240250) for "study-level", but one study may have more then 1 image, there are 2 typical scenarios that occur when multiple images are in a study(examples, www.kaggle.com/c/siim-covid19-detection/discussion/240878): 
    * The images are nearly identical (relatively common - differences are blurriness/sharpness/lighting/rotation)(examples in link) 
    * The images are different from each other (relatively uncommon)
    
2. Predict bbox(s) and his label(opacity/none) for "image-label"

*Example of data structure and his predicts for one study folder:*
```
123acsdf                
├── aksd34as          
│   └── ak123444.dcm   
├── k2765hgs
    └── a1233777.dcm  
```
* 123acsdf - study id
* ak123444, a1233777 - image ids

### Each image contain only 1 label?

````python
# Each image contain only 1 label? 

def count_only_unqiue_labels(string):
    string = re.sub(r"[^A-Za-z]", ' ', string)
    labelsnum = len(set(string.split()))
    return labelsnum

teststr1 = "opacity 1 101 101 200 200 opacity 0.4 100 100 200 200 none 1 0 0 1 1 flex 1 3 3 7"
print(f"UNIQUE LABELS NUM TEST: {count_only_unqiue_labels(teststr1)}")
    
labels = train_image_level_df['label']
labels_nums = labels.map(count_only_unqiue_labels)

cond_each_image_contain_only_1_label = all(labels_nums == 1)
assert cond_each_image_contain_only_1_label
print(f"Each image contain only 1 label: {cond_each_image_contain_only_1_label}")
````

### All bbox is NAN where label is none?

````python
# all bbox is NAN where label is none?

def get_label(string):
    string = re.sub(r"[^A-Za-z]", ' ', string)
    label = ' '.join(string.split())
    return label

teststr2 = "opacity 1 101 101 200 200"
print(f"LABEL EXTRACT TEST: {get_label(teststr2)}")

labels_boxes_df = train_image_level_df[['label', 'boxes']]
labels_boxes_df['label'] = labels_boxes_df['label'].map(get_label)

none_labels = labels_boxes_df[labels_boxes_df['label'] == 'none']
cond_all_bbox_is_NAN_where_label_is_none = all(none_labels['boxes'].isna())
assert cond_all_bbox_is_NAN_where_label_is_none
print(f"bbox is NAN where label is none: {cond_all_bbox_is_NAN_where_label_is_none}")
````

### All bbox exist where label is opacity?

````python
# all bbox exist where label is opacity?

exist_labels = labels_boxes_df[labels_boxes_df['label'] == 'opacity']
cond_all_bbox_exist_where_label_is_opacity = all(~exist_labels['boxes'].isna())
assert cond_all_bbox_exist_where_label_is_opacity
print(f"all bbox exist where label is opacity: {cond_all_bbox_exist_where_label_is_opacity}")
````

````python
# Label types

def get_unique_label(string):
    string = re.sub(r"[^A-Za-z]", ' ', string)
    label = ' '.join(set(string.split()))
    return label

teststr3 = "opacity 1 101 101 200 200 opacity 0.4 100 100 200 200 none 1 0 0 1 1 flex 1 3 3 7"
print(f"UNIQUE LABELS TEST: {get_unique_label(teststr3)}")

labels = train_image_level_df['label']
unique_labels = labels.map(get_unique_label)

print(f"Unique labels: {unique_labels.unique()}")
````

### Bboxes distribution on one image

````python
# bboxes distribution on one image

def count_labels(string):
    string = re.sub(r"[^A-Za-z]", ' ', string)
    labelsnum = len(string.split())
    return labelsnum

labels_labeltype_count = pd.DataFrame({
    'label': train_image_level_df['label'],
    'label_type': train_image_level_df['label'].map(get_unique_label),
    'bbox_count': train_image_level_df['label'].map(count_labels)
})

bbox_count = labels_labeltype_count[labels_labeltype_count['label_type'] == 'opacity']['bbox_count']
min_bboxnum,mean_bboxnum, max_bboxnum = bbox_count.min(), round(bbox_count.mean(), 2), bbox_count.max()

# plot 
_, ax = plt.subplots(figsize=(12, 5))
sns.histplot(bbox_count, ax=ax, stat='probability');
ax.set_title(f"Opacity distribution, min: {min_bboxnum} | mean: {mean_bboxnum} | max: {max_bboxnum}", fontsize=18);
ax.set_xlabel(f"Bbox count of one image", fontsize=15);
ax.set_ylabel(f"Frequency", fontsize=15);
````

## DICOM images / studies example

````python
# get study by id in format WITHOUT postfix "_study"
def get_study_by_id(study_id, df=train_study_level_df):
    study_id += '_study'
    return df[df['id'] == study_id]

# get image by id in format WITHOUT postfix "_image"
def get_image_by_id(image_id, df=train_image_level_df):
    image_id += '_image'
    return df[df['id'] == image_id]

# get images by StudyInstanceUID
def get_images_by_sid(sid, df=train_image_level_df):
    return df[df['StudyInstanceUID'] == sid]

# search image path by image id
def get_image_path_by_id(image_id, root_path=train_images_root_path):
    fn = image_id + '.dcm'
    tree = os.walk(root_path)
    for dir_path, _, file_names in tree:
        for f in file_names:
            if fn == f:
                return dir_path + '/' + fn
    return -1

def get_string_main_dicom_metadata(img):
    return f"""
        Modality: {img.Modality},
        Sex: {img.PatientSex},
        ImageType: {img.ImageType},
        ImagerPixelSpacing: {img.ImagerPixelSpacing},
        PhotometricInterpretation: {img.PhotometricInterpretation},
        Rows: {img.Rows},
        Columns: {img.Columns}
    """
    
def draw_bboxes(bbox_list, img):
    for bbox in bbox_list:
        x1, y1 = int(bbox['x']), int(bbox['y'])
        w, h = int(bbox['width']), int(bbox['height'])
        img = cv2.rectangle(img, (x1, y1), (x1+w, y1+h), (255, 0, 100), 7)
    return img
        
````

### Example 1. Negative for Pneumonia / full metadata

````python
# Example 1. Negative for Pneumonia 

# get dfs
study_negative = '005057b3f880'
study = get_study_by_id(study_negative)
images = get_images_by_sid(study_negative)
print(f"Study df: {study}")
print(f"\nImage df: {images}")

# get image id and his path
image_id = re.sub(r'\_image$', '', images['id'].iloc[0])
image_neg_path = get_image_path_by_id(image_id)

negative_dcm = read_file(image_neg_path, stop_before_pixels=False)
metadata_neg = get_string_main_dicom_metadata(negative_dcm)

# plot dicom image
_, ax = plt.subplots(figsize=(15, 9))
plt.imshow(negative_dcm.pixel_array, cmap=plt.cm.gray);
ax.set_title(f'Image with id: {image_id},  negative for pneumonia', fontsize=18);
ax.text(3500, 1200, metadata_neg, size=15, color='purple');
````

## sample_submission.csv

Study id's have postfix '_study' and his 'PredictionString' must have follow structure:
> LABEL PROBABILITY 0 0 1 1
1. __LABEL__. Must be one of this: negative(Negative for Pneumonia), typical(Typical Appearance), indeterminate(Indeterminate Appearance), atypical(Atypical Appearance).
2. __PROBABILITY__. This is probability of labeled desease.
3. Last part "__0 0 1 1__". Last part of PredictionString is constant, he same among all.


Image id's have postfix '_image' and his 'PredictionString' must have follow structure:
> LABEL PROBABILITY XMIN YMIN XMAX YMAX
1. __LABEL__. Must be one of two: opacity() and none.
2. __PROBABILITY__. This is probability of this bbox contain opacity. For 'none' label prob. must be 1.
3. Last part __XMIN YMIN XMAX YMAX__ its coordinates of bbox.
4. If image have several opacities, PredictionString must contain predict for each opacity(example, opacity 0.5 0 0 100 150 opacity 0.92 120 100 150 200)
